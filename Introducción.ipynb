**Ciclos**
---
**for:** has static allocated memory.
<br>
**while:** unknown memory needed, unknown amount of repetitions. Needs an end condition.

```python
#for loop
#sum from 1-100
def sum(n: int) -> None:
  x = 0
  for i in range(n+1):
    x += i
  print(x)
#while loop
#while num <=100
def sum2(n: int)-> None:
  x = 0 #val
  i = 0 #cont
  while(x+i) <= n: # could add '(x+i) <= n' as another conditional apart from x<= n but it is not necessary
    x+=i
    i+=1
  print(i)
  print(x)
```

```python
sum(100)
sum2(100)
```
